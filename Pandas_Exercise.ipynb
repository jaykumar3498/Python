```python
import numpy as np
import pandas as pd
```

```python
crime = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv')
crime.head()
```

```python
crime.shape
```

```python
crime.info()
```

```python
crime['Year'] = pd.to_datetime(crime['Year'], format='%Y')
crime.head()
```

```python
crime.info()
```

```python
crime.set_index('Year',drop=True,inplace=True)
```

```python
crime.head()
```

```python
crime.drop('Total',inplace=True,axis=1)
crime.head()
```

```python
crime2 = crime.resample('10AS').sum()
crime2
```

```python
crime2.iloc[:,1:].sum(axis=1).sort_values(ascending=False)
```
